### Kiểm tra phiên bản OpenCV


```python
from google.colab.patches import cv2_imshow
import numpy as np
import cv2
print(cv2.__version__)
```

### Đọc ảnh và các thuộc tính của ảnh


#### Đọc ảnh




```python
image_path = '/content/road_2.jpg' # Điền file ảnh cần đọc
# Đọc ảnh và trả về một ma trận 3 chiều ứng với các kênh màu B,G,R
img = cv2.imread(image_path)

img
```

### Thay đổi độ sáng của ảnh



#### Lời giải bài tập 1


```python
# Tạo một ma trận cùng kích thước với ảnh, với giá trị cộng thêm vào độ sáng
brightness_change = 50
matrix = np.ones(img.shape, dtype="uint8") * brightness_change

# Tăng độ sáng của ảnh bằng cách sử dụng hàm cv2.add
brightened_image = cv2.add(img, matrix)

# Giảm độ sáng của ảnh bằng cách sử dụng hàm cv2.subtract
darkened_image = cv2.subtract(img, matrix)
```

#### Kết quả


```python
# Tạo một figure với 3 subplot
fig, axs = plt.subplots(1, 3, figsize=(10, 5))

# Hiển thị ảnh đầu tiên
axs[0].imshow(img[:,:,::-1])
axs[0].axis('off')  # Tắt hiển thị trục

# Hiển thị ảnh thứ hai
axs[1].imshow(brightened_image[:,:,::-1])
axs[1].axis('off')  # Tắt hiển thị trục

# Hiển thị ảnh thứ ba
axs[2].imshow(darkened_image[:,:,::-1])
axs[2].axis('off')  # Tắt hiển thị trục

# Hiển thị figure
plt.show()
```

### Thay đổi kích thước ảnh

### Chuyển ảnh sang ảnh xám





#### Thực hành

```python
# Chuyển đổi sang ảnh xám bằng phương pháp trung bình cộng
gray_image_average = np.mean(img, axis=2).astype(np.uint8)
```

```python
# Chuyển đổi sang ảnh xám bằng OpenCV với công thức trọng số
gray_image_weighted = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

#### Kết quả


```python
# Tạo một figure với 2 subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Hiển thị ảnh thứ nhất
axs[0].imshow(gray_image_average, cmap = "gray")
axs[0].axis('off')  # Tắt hiển thị trục

# Hiển thị ảnh thứ hai
axs[1].imshow(gray_image_weighted, cmap = "gray")
axs[1].axis('off')  # Tắt hiển thị trục

# Hiển thị figure
plt.show()
```

### Chuyển ảnh sang các hệ màu khác (HLS và HSV)



#### Lời giải bài tập 4

```python
# Convert the input image to HLS color space
hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)

# Convert the input image to HSV color space
hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
```

#### Kết quả

```python
# Tạo một figure với 2 subplot
fig, axs = plt.subplots(1, 2, figsize=(10, 5))

# Hiển thị ảnh thứ nhất
axs[0].imshow(hls[:,:,::-1])
axs[0].axis('off')  # Tắt hiển thị trục

# Hiển thị ảnh thứ hai
axs[1].imshow(hsv[:,:,::-1])
axs[1].axis('off')  # Tắt hiển thị trục

# Hiển thị figure
plt.show()
```

### Các thao tác với video





#### Lời giải bài tập 6

```python
# Đường dẫn đến video đầu vào và video đầu ra
input_video_path = '/road.mp4'
output_video_path = 'output_video.mp4'

# Mở video đầu vào
cap = cv2.VideoCapture(input_video_path)

# Lấy thông tin về định dạng của video đầu vào
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# Tạo đối tượng VideoWriter để ghi video đầu ra
out = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height), isColor=False)

# Xác định các điểm đầu và điểm cuối của đường thẳng
start_point = (50, 50)  # Ví dụ: điểm đầu tại (50, 50)
end_point = (200, 200)  # Ví dụ: điểm cuối tại (200, 200)

# Màu sắc của đường thẳng (B, G, R) - ví dụ: màu trắng
color = 255  # Màu trắng trong ảnh xám
thickness = 2  # Độ dày của đường thẳng

# Giá trị giảm độ sáng
brightness_decrease = 50

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # Chuyển đổi khung hình sang ảnh xám
    gray_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    # Giảm độ sáng của ảnh
    darkened_frame = cv2.subtract(gray_frame, np.ones(gray_frame.shape, dtype="uint8") * brightness_decrease)

    # Vẽ đường thẳng lên ảnh
    cv2.line(darkened_frame, start_point, end_point, color, thickness)

    # Ghi khung hình đã xử lý vào video đầu ra
    out.write(darkened_frame)

# Giải phóng các tài nguyên
cap.release()
out.release()
cv2.destroyAllWindows()
```
